```python
import os
from PIL import Image
 
def convert_to_pixel_coordinates(x_center, y_center, width, height, img_width, img_height):
    x_center_abs = x_center * img_width
    y_center_abs = y_center * img_height
    w = width * img_width
    h = height * img_height
    x = int(x_center_abs - w / 2)
    y = int(y_center_abs - h / 2)
    return x, y, int(w), int(h)
 
txt_files_path = r"E:\SMU\Jan2024\CS604\GroupProject\LogoDetection\YOLO\runs\detect\result2"
image_files_path = r"E:\SMU\Jan2024\CS604\GroupProject\LogoDetection\eval_folder_"
output_file_path = r"E:\SMU\Jan2024\CS604\GroupProject\LogoDetection\object_est1.txt"
 
def find_image_file(image_file_name, image_files_path):
    for file in os.listdir(image_files_path):
        if file.lower() == image_file_name.lower():
            return os.path.join(image_files_path, file), file
    return None, None
 
with open(output_file_path, "w") as output_file:
    for txt_file in os.listdir(txt_files_path):
        if txt_file.endswith(".txt"):
            image_file_name_without_ext = txt_file[:-4]  # Remove ".txt" extension
            image_file_name_jpg = image_file_name_without_ext + ".jpg"
            img_file_path, actual_image_file_name = find_image_file(image_file_name_jpg, image_files_path)
            if img_file_path:
                with Image.open(img_file_path) as img:
                    img_width, img_height = img.size
                txt_file_path = os.path.join(txt_files_path, txt_file)
                with open(txt_file_path, "r") as file:
                    for line in file:
                        class_id, x_center, y_center, width, height = map(float, line.split())
                        x, y, w, h = convert_to_pixel_coordinates(x_center, y_center, width, height, img_width, img_height)
                        output_file.write(f"{actual_image_file_name}, {x}, {y}, {w}, {h}\n")
```
